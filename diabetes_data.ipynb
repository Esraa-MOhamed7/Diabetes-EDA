```python
for col in df.select_dtypes("object"):
    df[col]=df[col].astype("category")
df["BMI"]=df["BMI"].astype("float")
```

```python
for col in df.select_dtypes("int64"):
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    outliers=((df[col]<lower_bound) | (df[col]>upper_bound))
    print(f"{col} has {outliers} outliers")
```

```python
for col in df.select_dtypes("category"):
    print(df[col].value_counts())
```

```python
#add new features
#1 :
bins=[0,18.5,25,30,float('inf')]
labels=["Lack of Weight","Normal Weight","Weight gain","Obesity"]
df["BMI_groups"]=pd.cut(df["BMI"],bins=bins,labels=labels)
print(df["BMI_groups"].value_counts())
```

```python
#2 :
coditions=[(df["Family History"]=="Yes") & (df["Genetic Markers"]=="Yes"),
           (df["Family History"]=="Yes") | (df["Genetic Markers"]=="Yes"),
           (df["Family History"]=="No") & (df["Genetic Markers"]=="No")]
choices=["High Risk","Moderate Risk","Low Risk"]
df["genetic_family_risk"]=np.select(coditions,choices,default="UnKnown")
print(df["genetic_family_risk"].value_counts())
```

```python
#3 :
print(f"Blood Glucose Levels:{df["Blood Glucose Levels"].describe()}")
print(f"Insulin Levels:{df["Insulin Levels"].describe()}")
df["Blood Glucose Levels_mol"]=df["Blood Glucose Levels"]/18.018
df["HOMA_IR"]=(df["Insulin Levels"]*df["Blood Glucose Levels_mol"])/22.5
df["Insulin_Resistance"]=np.where(df["HOMA_IR"]>2.5,"Resistance","Sensitive")
print(df[["Blood Glucose Levels","Blood Glucose Levels_mol","Insulin Levels","HOMA_IR"]].head())
```

```python
#4 :
bins=[0,18,54,65,float("inf")]
labels=["Child/Adolescent","Young Adult","Middle-Agged","Senior"]
df["age_group"]=pd.cut(df["Age"],bins=bins,labels=labels)
print(df["age_group"].value_counts())
```

Analysis of Insulin Levels for each Type of Daibetes

```python
plt.figure(figsize=(12,6))
sns.set_palette("Pastel2")
sns.boxplot(data=df,x="Target",y="Insulin Levels")
plt.title("Distribution of Insulin Levels for each Type of Daibetes")
plt.xlabel("Type of Diabetes")
plt.ylabel("Insulin Levels")
plt.xticks(rotation=45,ha="center")
plt.tight_layout()
plt.show()
```

Analysis of the Relationship between Insulin Levels and Blood Pressure

```python
plt.figure(figsize=(12,6))
sns.set_palette("dark:#5A9_r")

sns.scatterplot(data=df,x="Insulin Levels",y="Blood Pressure",hue="Target")
plt.title("The Relationship between Insulin Levels and Blood Pressure")
plt.xlabel("Insulin Levels")
plt.ylabel("Blood Pressure")
plt.legend(title="Type of Diabetes",title_fontsize=8,fontsize=9,bbox_to_anchor=(1.05, 1),loc='upper left')
plt.tight_layout()
plt.show()
```

Analysis of the Relationship between Types of Diabetes and Behavioral Factors

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(28, 6))
sns.set_palette("crest")

# Physical Activity
activity_target = pd.crosstab(df["Physical Activity"], df["Target"])
activity_target.plot(kind="bar", ax=axes[0])
axes[0].set_title("Distribution of Physical Activity by Type of Diabetes", fontsize=14)
axes[0].set_xlabel("Level of Physical Activity", fontsize=12)
axes[0].set_ylabel("Number of Cases", fontsize=12)
axes[0].tick_params(axis="x", rotation=45)
axes[0].legend(title="Type of Diabetes", title_fontsize=11, fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')

# Dietary Habits
diet_target = pd.crosstab(df["Dietary Habits"], df["Target"])
diet_target.plot(kind="bar", ax=axes[1])
axes[1].set_title("Distribution of Dietary Habits by Type of Diabetes", fontsize=14)
axes[1].set_xlabel("Dietary Habits", fontsize=12)
axes[1].set_ylabel("Number of Cases", fontsize=12)
axes[1].tick_params(axis="x", rotation=45)
axes[1].legend(title="Type of Diabetes", title_fontsize=11, fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')

# Smoking Status
smoking_target = pd.crosstab(df["Smoking Status"], df["Target"])
smoking_target.plot(kind="bar", ax=axes[2])
axes[2].set_title("Distribution of Smoking Status by Type of Diabetes", fontsize=14)
axes[2].set_xlabel("Smoking Status", fontsize=12)
axes[2].set_ylabel("Number of Cases", fontsize=12)
axes[2].tick_params(axis="x", rotation=45)
axes[2].legend(title="Type of Diabetes", title_fontsize=11, fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

Analysis of the Prevalence of Types of Diabetes for each Age Group

```python
age_target_count = pd.crosstab(df["age_group"], df["Target"])
sns.set_palette("dark:#5A9_r")
age_target_count.plot(kind="bar", figsize=(10,6), stacked=True)

plt.title("Prevalence of Types of Diabetes for Each Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Cases")
plt.legend(title="Type of Diabetes",title_fontsize=8,fontsize=9,bbox_to_anchor=(1.05, 1),loc='upper left')
plt.tight_layout()
plt.show()
```

What is the Average BMI for each Type og Diabetes?

```python
avg_bmi_per_target=df.groupby("Target",observed=True)["BMI"].mean()
sns.set_palette("Pastel2")
avg_bmi_per_target.plot(kind="bar",figsize=(10,6))
plt.title("Average BMI for each Type og Diabetes")
plt.xlabel("Type of Diabetes")
plt.ylabel("Average of BMI")
plt.tight_layout()
plt.show()
```
